```python
def valid(grid, row, col, num):

    for x in range(9):
        if grid[row][x] == num:
            return False

    for x in range(9):
        if grid[x][col] == num:
            return False
        
    subgrid_row = (row // 3) * 3
    subgrid_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if grid[i + subgrid_row][j + subgrid_col] == num:
                return False
    return True

def solveSudoku(grid, row, col):

    if (row == 8 and col == 9):
        return True

    if col == 9:
        row += 1
        col = 0

    if grid[row][col] > 0:
        return solveSudoku(grid, row, col + 1)
    for num in range(1, 10, 1):
        if valid(grid, row, col, num):
            grid[row][col] = num
            if solveSudoku(grid, row, col + 1):
                return True
        grid[row][col] = 0
    return False

def print_grid(grid):
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - - - - ")
        for j in range(9):
            if j % 3 == 0 and j != 0:
                print(" | ", end="")
            if j == 8:
                print(grid[i][j])
            else:
                print(str(grid[i][j]) + " ", end="")
grid=[[4, 3, 0, 0, 0, 0, 2, 0, 0], [5, 0, 1, 4, 2, 0, 0, 0, 7], [0, 0, 0, 1, 7, 0, 0, 8, 0], [0, 0, 0, 0, 0, 0, 5, 0, 0], [0, 0, 0, 0, 0, 3, 0, 2, 0], [2, 6, 0, 5, 0, 1, 4, 7, 9], [0, 0, 0, 3, 9, 0, 7, 5, 1], [9, 0, 0, 6, 0, 0, 0, 0, 3], [3, 2, 0, 0, 1, 0, 6, 9, 4]]
print_grid(grid)
if (solveSudoku(grid, 0, 0)):
    print("Solution of the Sudoku Puzzle")
    print_grid(grid)
else:
    
    print("There is no solution ")
 
```
