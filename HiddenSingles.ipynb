```python
# Sudoku puzzle to solve
puzzle = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

# Print the puzzle
def print_puzzle(puzzle):
    for row in puzzle:
        print(row)

# Check if a value is valid for a given position
def is_valid(puzzle, row, col, value):
    # Check row
    for i in range(9):
        if puzzle[row][i] == value:
            return False

    # Check column
    for i in range(9):
        if puzzle[i][col] == value:
            return False

    # Check 3x3 box
    box_row = (row // 3) * 3
    box_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if puzzle[box_row + i][box_col + j] == value:
                return False

    # If all checks pass, the value is valid
    return True

# Find all possible values for an empty cell
def possible_values(puzzle, row, col):
    values = []
    for i in range(1, 10):
        if is_valid(puzzle, row, col, i):
            values.append(i)
    return values

# Find all hidden singles in the puzzle
def find_hidden_singles(puzzle):
    for row in range(9):
        for col in range(9):
            if puzzle[row][col] == 0:
                values = possible_values(puzzle, row, col)
                if len(values) == 1:
                    puzzle[row][col] = values[0]
                    print(f"Found hidden single ({values[0]}) at ({row}, {col})")
                    return True
    return False

def print_grid(grid):
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - - - - ")
        for j in range(9):
            if j % 3 == 0 and j != 0:
                print(" | ", end="")
            if j == 8:
                print(grid[i][j])
            else:
                print(str(grid[i][j]) + " ", end="")
                
# Solve the puzzle using the hidden singles method
def solve_hidden_singles(puzzle):
    while find_hidden_singles(puzzle):
        pass

# Solve the puzzle
solve_hidden_singles(puzzle)

# Print the solution
print_grid(puzzle)
```
